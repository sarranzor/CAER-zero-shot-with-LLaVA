# Baseline ResNet — Evaluation

## 1. Imports & Config

```python
import json
from pathlib import Path

from dual_resnet_shared import (
    device, EMOTIC_LABELS, NCAERS_LABELS,
    make_emotic_loaders, make_ncaers_loaders,
    make_synth_loaders, build_synth_index,
)
from baseline_model import BaselineModel, evaluate_model, plot_training_curves

CKPT_DIR = Path("checkpoints/checkpoints/resnet_baseline")

print(f"Device: {device}")
```

## 2. EMOTIC

```python
_, _, emotic_test_loader, _ = make_emotic_loaders()

emotic_model = BaselineModel(freeze_backbone=False).to(device)
emotic_model.load_state_dict(__import__("torch").load(CKPT_DIR / "emotic_best.pt", map_location=device))

with open(CKPT_DIR / "emotic_history.json") as f:
    emotic_history = json.load(f)

print("EMOTIC loaded.")
```

```python
emotic_metrics = evaluate_model(
    emotic_model, emotic_test_loader,
    "EMOTIC", EMOTIC_LABELS, "EMOTIC",
    CKPT_DIR / "emotic_bs",
)
```

```python
plot_training_curves(emotic_history, "EMOTIC", EMOTIC_LABELS, emotic_metrics,
                     CKPT_DIR / "emotic_results_bs.png")
```

## 3. NCAERS

```python
_, _, ncaers_test_loader, _, _ = make_ncaers_loaders()

ncaers_model = BaselineModel(freeze_backbone=False).to(device)
ncaers_model.load_state_dict(__import__("torch").load(CKPT_DIR / "ncaers_best.pt", map_location=device))

with open(CKPT_DIR / "ncaers_history.json") as f:
    ncaers_history = json.load(f)

print("NCAER-S loaded.")
```

```python
ncaers_metrics = evaluate_model(
    ncaers_model, ncaers_test_loader,
    "CAER_S", NCAERS_LABELS, "NCAERS",
    CKPT_DIR / "ncaers_bs",
)
```

```python
plot_training_curves(ncaers_history, "NCAERS", NCAERS_LABELS, ncaers_metrics,
                     CKPT_DIR / "ncaers_results_bs.png")
```
